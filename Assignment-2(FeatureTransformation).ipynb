```python
import pandas as pd
```

```python
# read and store new dataset
data1 = pd.read_csv('supershops.csv')
data1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler # importing MinMaxScaler to transform data
```

```python
mms = MinMaxScaler() # Creating object of MinMaxScaler
```

```python
# Transforming data by fit_transform function which will fit first then transfrom tha data
data1['Marketing Spend'] = mms.fit_transform(data1[['Marketing Spend']])
data1['Administration'] = mms.fit_transform(data1[['Administration']])
data1['Transport'] = mms.fit_transform(data1[['Transport']])
data1['Profit'] = mms.fit_transform(data1[['Profit']])
```

```python
data1.head() # data after transform
```
